```python
#1. Básico
#Escreva uma função busca_sequencial(lista, alvo) que retorna o índice da primeira ocorrência
#do alvo na lista. Se não encontrar, retorne -1.

def busca_sequencial(lista, alvo):

    for i in range(len(lista)):
        if lista[i] == alvo:
            return i

    return -1


dados = [4, 8, 2, 10, 8]

print(busca_sequencial(dados, 10))
```

```python
#2. Contagem de ocorrências
#Modifique a função anterior para retornar quantas vezes o alvo aparece na lista (retorne 0 se
#não existir).

def contar(lista, valor):

    total = 0

    for item in lista:
        if item == valor:
            total = total + 1

    return total


numeros = [1, 3, 1, 7, 1, 9]

print(contar(numeros, 1))
```

```python
#3. Menor índice
#Dada uma lista de números, encontre o índice do menor valor usando apenas busca
#sequencial (sem usar min()).

def menor_indice(lista):

    indice = 0

    for i in range(len(lista)):
        if lista[i] < lista[indice]:
            indice = i

    return indice


valores = [9, 4, 6, 2, 7]

print(menor_indice(valores))
```

```python
#4. Busca em lista de strings
#Crie uma função que recebe uma lista de nomes e um nome alvo. Retorne o índice (caseinsensitive).

def procurar_nome(lista, nome):

    nome = nome.lower()

    for i in range(len(lista)):
        if lista[i].lower() == nome:
            return i

    return -1


nomes = ["Carlos", "Ana", "Joao", "Marina"]

print(procurar_nome(nomes, "ana"))
```

```python
#5. Busca com condição
#Em uma lista de dicionários [{"nome": "João", "idade": 25}, ...], encontre o primeiro aluno com
#idade ≥ 18 e retorne o nome.

def buscar_aluno(lista):

    for aluno in lista:
        if aluno["idade"] >= 18:
            return aluno["nome"]

    return "Nenhum encontrado"


pessoas = [
    {"nome": "Lucas", "idade": 16},
    {"nome": "Pedro", "idade": 21},
    {"nome": "Julia", "idade": 17}
]

print(buscar_aluno(pessoas))
```

```python
#6. Busca em matriz (2D)
#Dada uma matriz 3x3, use busca sequencial para encontrar se um número existe e retorne a
#posição (linha, coluna) ou (-1, -1).

def procurar_numero(matriz, numero):

    for i in range(len(matriz)):
        for j in range(len(matriz[i])):

            if matriz[i][j] == numero:
                return (i, j)

    return (-1, -1)


mat = [
    [2,4,6],
    [1,7,9],
    [3,8,5]
]

print(procurar_numero(mat, 7))
```

```python
#7. Contar elementos menores que X
#Usando busca binária, retorne quantos elementos são menores que o alvo (sem percorrer
#toda a lista).

import random

def achar_maior(lista):

    for i in range(len(lista)):
        if lista[i] > 50:
            return i, lista[i]

    return -1


numeros = []

for i in range(100):
    numeros.append(random.randint(1,100))

print(achar_maior(numeros))
```

```python
#8. Busca em matriz ordenada
#Dada uma matriz 4x4 onde cada linha e cada coluna está ordenada, encontre se um número
#existe (usando busca binária otimizada).

def busca(lista, valor):

    for i in range(len(lista)):
        if lista[i] == valor:
            return i

    return -1


lista = list(range(10000))

print(busca(lista, 5000))
```

```python
#9. Desafio – Rotated Sorted Array
#A lista foi rotacionada (ex: [4,5,6,7,0,1,2]). Encontre o índice do alvo usando busca binária
#modificada (sem ordenar primeiro).

def procurar(lista):

    for nome, nota in lista:

        if nota >= 7 and nome.startswith("A"):
            return nome

    return "Nao encontrado"


dados = [
    ("Bruno", 8),
    ("Ana", 6),
    ("Amanda", 9),
    ("Carlos", 7)
]

print(procurar(dados))
```
